```python
#import file

Zip_df = pd.read_csv("Zip_Zhvi_AllHomes.csv", encoding = 'latin-1')
```

```python
Zip_df.head()
```

```python
#List of cities we want

county_name = ["Philadelphia County"]
#Zip_df["City"].isin(cities_list)

Zip_df = Zip_df.loc[Zip_df["CountyName"].isin(county_name)]
Zip_df.head()
```

```python
#Drop unwanted columns
#reset index to 0

drop_column = ["State" , "RegionID" , "Metro" , "City" , "CountyName" , "SizeRank"]
Zip_trimmed = Zip_df.drop(drop_column, axis = 1)

Zip_trimmed
home_values = Zip_trimmed.reset_index(drop = True)
home_values.head()
```

```python
#Melting data to make it in tidy vertcial format

ZHVI = pd.melt(
    home_values, id_vars=["RegionName"], value_name="ZHVI", var_name="Date"
)
ZHVI
```

```python
# convert to timestamp
ZHVI['Date'] = pd.to_datetime(ZHVI['Date'])
```

```python
ZHVI['Month'] = ZHVI['Date'].dt.month
ZHVI['Year'] = ZHVI['Date'].dt.year
```

```python
ZHVI
```

```python
annual_ZHVI = ZHVI.groupby(["RegionName", "Year"])["ZHVI"].mean()
annual_ZHVI.head()
```

```python
annual_ZHVI = annual_ZHVI.reset_index()
```

```python
annual_ZHVI
```

```python
#sorting the ZHVI value in descending order

annual_ZHVI_sorted = annual_ZHVI.sort_values(by = ["ZHVI"] , ascending = False)
annual_ZHVI_sorted
```
